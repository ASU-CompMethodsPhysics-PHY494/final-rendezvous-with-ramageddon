```python
import rama
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
#%matplotlib inline
%matplotlib notebook
```

```python
r = np.ones((3,3)) #units of AU
r[0] = [0,0,0] #sun
#r[1] = [0.38709927,0,0] #mercury
#r[2] = [0.72333566,0,0] #venus
#r[3] = [1.00000261,0,0] #earth
r[1] = [19.1914,0,0] #uranus
r[2] = [30.0611,0,0] #neptune

masses = np.ones(3) #units of solar masses
masses[0] = 1.
#masses[1] = 1/6023600 #mercury
#masses[2] = 1/408523.71 #venus
#masses[3] = 1/328900.56 #earth
masses[1] = 4.366244e-5 #uranus
masses[2] = 5.151389e-5 #neptune

v = np.zeros((3,3)) #units of AU per year
v[0] = [0,0,0] #sun
#v[1] = [0,0,365.256*2*np.pi*(r[1][1])/87.969] #mercury
#v[2] = [0,0,365.256*2*np.pi*(r[2][1])/224.701] #venus
#v[3] = [0,0,365.256*2*np.pi*(r[3][1])/365.256] #earth
v[1] = [0,0,2*np.pi*19.1914 / 84.0110] #uranus
v[2] = [0,0,2*np.pi*30.0611 / 164.7901] #neptune

#v = -365.256*v
```

```python
tmax = 240 #units of years
dt = 0.01 #units of years
t = np.linspace(0, tmax, int(tmax/dt))

xt, vt, KE, UE, TE = rama.dynamics(r, v, dt, masses, tmax)
```

```python
fig2 = plt.figure()
ax2 = fig2.add_subplot(111, projection='3d')
ax2.scatter(xt[:,0,0],xt[:,0,1],xt[:,0,2], color='red',)
ax2.plot(xt[:,1,0],xt[:,1,1],xt[:,1,2], color='blue',)
ax2.plot(xt[:,2,0],xt[:,2,1],xt[:,2,2], color='green',)
#ax2.plot(xt[:,3,0],xt[:,3,1],xt[:,3,2], color='black',)
```
