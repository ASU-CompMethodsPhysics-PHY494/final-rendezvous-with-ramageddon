```python
import rama
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
#%matplotlib inline
%matplotlib notebook
```

```python
r = np.ones((3,3))
r[0] = [0,0,0]
r[1] = [19.1914,0,0]
r[2] = [30.0611,0,0]
masses = np.ones(3)
masses[0] = 1
masses[1] = 4.366244e-5
masses [2] = 5.151389e-5
v = np.ones((3,3))
v[1] = [0,-1e-6,0]
v[2] = [0,-1e-6,0]
v[0] = [0,0,0]
```

```python
force, u = rama.gravity(r,masses)
```

```python
nsteps = 10000

xt, vt, KE, UE, TE = rama.dynamics(r, v, 0.001, masses, nsteps)
```

```python
fig2 = plt.figure()
ax2 = fig2.add_subplot(111, projection='3d')
ax2.scatter(xt[:,0,0],xt[:,0,1],xt[:,0,2], color='red',)
ax2.plot(xt[:,1,0],xt[:,1,1],xt[:,1,2], color='blue',)
ax2.plot(xt[:,2,0],xt[:,2,1],xt[:,2,2], color='green',)
```
